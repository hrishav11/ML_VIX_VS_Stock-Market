```python
# Converts the “observation_date” column from string to datetime format
data_vix['observation_date'] = pd.to_datetime(data_vix['observation_date'])
data_nasdq['observation_date'] = pd.to_datetime(data_nasdq['observation_date'])
```

```python
# This code sorts the dataset by date in ascending order.
data_vix = data_vix.sort_values('observation_date')
data_nasdaq = data_nasdq.sort_values('observation_date')
```

```python
# Merging the datasets
import pandas as pd
import numpy as np

# ----------------------------------------------------
# 1. Convert dates (your datasets use DD/MM/YYYY)
# ----------------------------------------------------
data_vix['observation_date'] = pd.to_datetime(data_vix['observation_date'], dayfirst=True)
data_nasdq['observation_date'] = pd.to_datetime(data_nasdq['observation_date'], dayfirst=True)

# ----------------------------------------------------
# 2. Sort by date
# ----------------------------------------------------
data_vix = data_vix.sort_values('observation_date')
data_nasdq = data_nasdq.sort_values('observation_date')

# ----------------------------------------------------
# 3. Filter the date range
# ----------------------------------------------------
start_date = "2000-01-03"
end_date = "2025-11-25"

data_vix_filtered = data_vix[
    (data_vix['observation_date'] >= start_date) &
    (data_vix['observation_date'] <= end_date)
]

data_nasdaq_filtered = data_nasdq[
    (data_nasdq['observation_date'] >= start_date) &
    (data_nasdq['observation_date'] <= end_date)
]

# ----------------------------------------------------
# 4. Merge the filtered datasets
# ----------------------------------------------------
data_Nasdaq_VIX_Merged = pd.merge(
    data_vix_filtered,
    data_nasdaq_filtered,
    on='observation_date',
    how='inner',
    suffixes=('_VIX', '_NASDAQ')
)

# ----------------------------------------------------
# 5. Compute daily log returns (creates last column)
# ----------------------------------------------------
data_Nasdaq_VIX_Merged["nasdaq_ret"] = np.log(
    data_Nasdaq_VIX_Merged["NASDAQ100"] /
    data_Nasdaq_VIX_Merged["NASDAQ100"].shift(1)
)

# ----------------------------------------------------
# 6. Fill the first log return manually (no dropping)
# ----------------------------------------------------
data_Nasdaq_VIX_Merged.iloc[0, data_Nasdaq_VIX_Merged.columns.get_loc("nasdaq_ret")] = -0.0610

# ----------------------------------------------------
# 7. Convert back to DD/MM/YYYY for your output
# ----------------------------------------------------
data_Nasdaq_VIX_Merged["observation_date"] = data_Nasdaq_VIX_Merged[
    "observation_date"
].dt.strftime("%d/%m/%Y")

# Final preview
data_Nasdaq_VIX_Merged.head()
```

FOR BASIC UNDERSTANDING

VIX --> market fear/uncertainty index for the US stock market

High VIX → market expects big moves (often in a crisis, lots of fear).

Low VIX → market expects calm, smaller moves.

It can be inferred from this graph that while Nasdaq’s daily log returns oscillate around zero with many random positive excursions, VIX returns manifest themselves with more sustained cycles interspersed with abrupt peaks at times of intense market stress episodes. Most notably, these periods of major VIX excursions—2001 dot-com bubble burst, 2008 global financial crisis, 2011 debt-ceiling crisis episode, 2020 COVID panic crash, and 2022 inflation scare—are exactly when huge adverse Nasdaq market returns manifest themselves. Concomitant with this indication of an established inverse link between market fear (suggested by high VIX) and stock market activity (manifested by more extreme and heightened volatile market excursions of Nasdaq), while lack of daily pair correlation can easily be inferred visually from this graph: periods of extreme VIX excursions display intense association with major market sell-offs and therefore can reliably function as an indicator not of market outcomes but rather market stress episodes themselves.


```python
# -----------------------------
# 1. Load and clean data
# -----------------------------
import pandas as pd
import numpy as np
import plotly.express as px

# Load your cleaned CSV
data = pd.read_csv('data_Nasdaq_VIX_Merged.csv')

# Fix date parsing automatically – handles mixed formats
data['observation_date'] = pd.to_datetime(data['observation_date'], format='mixed', dayfirst=True)

# Sort
data = data.sort_values('observation_date')

# -----------------------------
# 2. Compute daily volatility (absolute log return)
# -----------------------------
data['volatility'] = data['nasdaq_ret'].abs()

# -----------------------------
# 3. Animated interactive plot
# -----------------------------
fig = px.scatter(
    data,
    x="VIXCLS",
    y="volatility",
    animation_frame=data['observation_date'].dt.year.astype(str),
    animation_group="observation_date",
    size="volatility",
    color="VIXCLS",
    hover_name="observation_date",
    labels={
        "VIXCLS": "VIX (Market Fear Index)",
        "volatility": "NASDAQ-100 Daily Volatility",
    },
    title="How Market Fear (VIX) Influences NASDAQ-100 Volatility Over Time",
    height=600
)

fig.update_layout(
    coloraxis_colorbar=dict(title="VIX Level"),
    xaxis=dict(title="VIX (Fear Index)"),
    yaxis=dict(title="NASDAQ-100 Daily Volatility (|log return|)")
)

fig.show()
print(data_Nasdaq_VIX_Merged.head())
```
